# Python Plotting

A lot of the basic plotting in Python is done through a package called matplotlib. It is desgined to resemble the plotting functionality and syntax of MATLAB, so many of the functions reviewed below are very similar to the MATLAB tutorial

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
b2 = np.arange(2,8,0.5)
b3 = np.linspace(0.4,8.2,10)
```

```python
f1=b2**2
plt.plot(b2,f1)
```

```python
plt.plot(b2, b2**2, 'o', color='green',label = "$x^2$")
# plotting the line 2 points 
plt.plot(b2, b2**3, 'o', color='red',label = "$x^3$")
plt.xlabel('x - axis',fontsize=18)
# Set the y axis label of the current axis.
plt.ylabel('y - axis',fontsize=18)
# Set a title of the current axes.
plt.title('Two or more lines on same plot with suitable legends ')
# show a legend on the plot
plt.legend()
# Display a figure.
plt.show()
```

```python
[B2,B3] = np.meshgrid(b2,b3)
F3 = np.sin(B2) + np.exp(-(B3-4)**2)
plt.contour(B2,B3,F3)
plt.colorbar          
plt.xlabel('B2',fontsize=20)
plt.ylabel('B3',fontsize=20)
plt.title('F3(B2,B3)',fontsize=20)
```

```python
plt.contourf(B2,B3,F3)
plt.colorbar         
plt.xlabel('B2',fontsize=20)
plt.ylabel('B3',fontsize=20)
plt.title('F3(B2,B3)',fontsize=20)
plt.colorbar()
```

```python
plt.pcolor(B2,B3,F3)
plt.colorbar         
plt.xlabel('B2',fontsize=20)
plt.ylabel('B3',fontsize=20)
plt.title('F3(B2,B3)',fontsize=20)
plt.colorbar()
#default colormap in matplotlib is perceptually uniform "viridis" - is seen and interpreted the same regardless of color blindness
```

```python
plt.pcolor(B2,B3,F3,cmap='Greys')
plt.colorbar         
plt.xlabel('B2',fontsize=20)
plt.ylabel('B3',fontsize=20)
plt.title('F3(B2,B3)',fontsize=20)
plt.colorbar()
```

```python
plt.pcolor(B2,B3,F3,cmap='seismic',vmin=-1,vmax=1)
plt.colorbar         
plt.xlabel('B2',fontsize=20)
plt.ylabel('B3',fontsize=20)
plt.title('F3(B2,B3)',fontsize=20)
plt.colorbar()
```

```python
plt.pcolor(B2,B3,F3,cmap='hsv')
plt.colorbar         
plt.xlabel('B2',fontsize=20)
plt.ylabel('B3',fontsize=20)
plt.title('F3(B2,B3)',fontsize=20)
plt.colorbar()
```

```python
import matplotlib.colors as colors
F4 = np.exp(3*F3)
plt.pcolor(B2,B3,F4,cmap='Greys',norm=colors.LogNorm(vmin=F4.min(), vmax=F4.max()))
plt.colorbar         
plt.xlabel('B2',fontsize=20)
plt.ylabel('B3',fontsize=20)
plt.title('F3(B2,B3)',fontsize=20)
plt.colorbar()
```

```python
from mpl_toolkits import mplot3d
ax = plt.axes(projection='3d')
ax.plot_surface(B2,B3,F3, rstride=1, cstride=1,cmap='viridis', edgecolor='none')
ax.set_title('surface');
plt.colorbar         
plt.xlabel('B2',fontsize=20)
plt.ylabel('B3',fontsize=20)
plt.title('F3(B2,B3)',fontsize=20)
```

```python
X = 0.3*np.random.randn(10,12) + F3
```

```python
n, bins, patches = plt.hist(X.flatten(), bins=10)
```

```python
plt.plot((bins[0:10]+bins[1:11])/2,n,color='blue',linewidth=3)
plt.xlabel('X values',fontsize=20)
plt.ylabel('frequency',fontsize=20)
```

```python
plt.boxplot(X[:,7])
```

```python
plt.boxplot(X, positions=b2)
plt.xlabel('b2',fontsize=20)
plt.ylabel('X',fontsize=20)
```
